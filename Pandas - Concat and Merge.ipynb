```python
import pandas as pd
data1 = pd.DataFrame({'key1': ['K0','K1', 'K2', 'K3'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 

data1
```

```python
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

data2
```

```python
data3 = data1.merge(data2, how='left', left_on='key1', right_on='key1')

data3
```

```python
data4 = pd.concat([data1,data2], axis=0, ignore_index=True)

data4
```
