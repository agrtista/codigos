```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.DataFrame({'Animal': ['Pato','Gallo','Cerdo','Gallo','Pato'],
                        'Cantidad': [1,4,2,3,5]})

datos
```

```python
table = pd.pivot_table(datos, values='Cantidad', index=['Animal'],
                    #columns=['Animal'], 
                       aggfunc=np.sum)
table
```

```python
plt.plot(table)
plt.show()
```
